```python
import re
```

```python
def dms2dd(degrees, minutes, seconds):
    dd = float(degrees) + float(minutes)/60 + float(seconds)/(60*60);
#     if direction == 'E' or direction == 'N':
#         dd *= -1
    return dd;

def parse_dms(dms):
    parts = re.split('[˚\'"]+', dms)
    #print(parts)
    lat = dms2dd(parts[0], parts[1], parts[2])

    return (lat)

def parse_dms2(dms):
    parts = re.split('[⁰\'"]+', dms)
    #print(parts)
    lat = dms2dd(parts[0], parts[1], parts[2])

    return (lat)
```

```python
# foute data, cleaning
geolocations = geolocations.drop([27])
geolocations = geolocations.drop([40])
geolocations = geolocations.drop([69])
geolocations = geolocations.drop([84])
geolocations = geolocations.drop([98])
geolocations = geolocations.drop([141])
geolocations = geolocations.drop([152])
geolocations = geolocations.drop([173])
geolocations = geolocations.drop([233])
```

```python
teller = 0
LongitudeS = []
LatitudeS = []
for i in range(len(geolocations['P1_Longitude'])):
    teller += 1
    if '⁰' in geolocations['P1_Longitude'].iloc[i]:
        long = parse_dms2(geolocations['P1_Longitude'].iloc[i])
        lat = parse_dms2(geolocations['P1_Latitude'].iloc[i])
        LongitudeS.append(long)
        LatitudeS.append(lat)
        print(teller)
        print(long, lat)
    else:
        long = parse_dms(geolocations['P1_Longitude'].iloc[i])
        lat = parse_dms(geolocations['P1_Latitude'].iloc[i])
        LongitudeS.append(long)
        LatitudeS.append(lat)
        print(teller)
        print(long, lat)
```

```python
data = {'Longitude':  LongitudeS,
        'Latitude': LatitudeS
        }
Converted = pd.DataFrame (data, columns = ['Longitude','Latitude'])
Converted
```
